### set up class to extract news headlines and evaluate their sentiment

```python
from textblob import TextBlob
import requests
from bs4 import BeautifulSoup
import nltk
from nltk.corpus import stopwords
import numpy as np
import pandas as pd

class NewsAnalysis:
    def __init__(self, term, text_type):
        self.term = term
        self.text_type = text_type
        self.subjectivity = 0
        self.polarity = 0
        
        self.url = 'https://www.google.com/search?q={0}&source=lnms&tbm=nws&sa=X&ved=0ahUKEwiXnsS15cDgAhWC5oMKHdUyBL4Q_AUIDigB'.format(self.term)
        response = requests.get(self.url)
        self.soup = BeautifulSoup(response.text, 'html.parser') 
        
    def get_text(self): 
        if self.text_type == 'headline':
            results = self.soup.find_all(class_='r')
        elif self.text_type == 'subline':
            results = self.soup.find_all('div',class_='st')
        
        result_list = []

        for h in results:
            result_list.append(h.get_text())
        self.text = result_list
        
    def get_sentiment(self):
        for t in self.text:
            blob = TextBlob(t)
            self.subjectivity += blob.sentiment.subjectivity
            self.polarity += blob.sentiment.polarity
            
        print("Subjectivty: ", self.subjectivity / len(self.text))
        print("Polarity: ", self.polarity / len(self.text))
        
    def get_freq_words(self):
        stop_words = set(stopwords.words('english'))
        word_list = []
        for s in self.text:
            for word in s.split():
                if not word in stop_words:
                    word_list.append(word)

        freq_words = nltk.FreqDist(word_list)
        word_list_df = pd.DataFrame(word_list)
        return(word_list_df[0].value_counts())
        
    def get_positive(self):
        pos_text = []
        for t in self.text:
            blob = TextBlob(t)
            if blob.sentiment.polarity > 0:
                pos_text.append(t)
        return pos_text
    
    def get_negative(self):
        pos_text = []
        for t in self.text:
            blob = TextBlob(t)
            if blob.sentiment.polarity < 0:
                pos_text.append(t)
        return pos_text
```

### look at headlines

### generate wordcloud for positive headlines

```python
from wordcloud import WordCloud, STOPWORDS 
import matplotlib.pyplot as plt 
stopwords = set(STOPWORDS)


positive = a.get_positive()

wordcloud = WordCloud(width = 800, height = 800, 
                background_color ='white', 
                stopwords = stopwords,
                min_font_size = 10).generate(str(positive))
  
# plot the WordCloud image                        
plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 
  
plt.show() 
```
